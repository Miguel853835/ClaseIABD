## Varios Gráficos en un Plot

```python
import numpy as np
import matplotlib.pyplot as pl

# Distribución 1
distribucion1X = np.array([25, 36, 14, 32, 18])
distribucion1Y = np.array([10, 50, 20, 40, 30])
pl.plot(distribucion1X, distribucion1Y, "y*", label ="Distribución 1")

# Distribución 2
distribucion2X = np.array([31, 14, 21, 18, 11])
distribucion2Y = np.array([25, 35, 25, 15, 55])
pl.plot(distribucion2X, distribucion2Y, color = 'r', marker= 'o', linestyle = '--', label ="Distribución 2")

# Distribución 3
distribucion3X = np.array([15, 25, 30, 35, 40])
distribucion3Y = np.array([12, 20, 25, 27, 29])
pl.plot(distribucion3X, distribucion3Y, '--go', markersize = 15, linewidth = 3, label ="Distribución 3")

pl.legend()
pl.xlabel("Coordenada X")
pl.ylabel("Coordenada Y")
pl.xlim(0, 50)
pl.ylim(0, 60)
pl.show()
```

Podemos cambiar el color o la forma en que se unen los puntos. Hay varios colores y marcadores ya definidos:

* 'b' as blue
* 'g' as green
* 'r' as red
* 'c' as cyan
* 'm' as magenta
* 'y' as yellow
* 'k' as black
* 'w' as white

Aunque se podría especificar cualquier color en concreto usando distintas escalas (https://matplotlib.org/stable/tutorials/colors/colors.html). 

Marcadores también hay muchos, los más comunes son:

* '.'  puntos
* 'o'  circulo
* '^'  triángulo
* 'sq' cuadrado
* '--' línea discontinua
* '..' línea dashed

Podéis verlos todos [aquí](https://matplotlib.org/stable/api/markers_api.html). Tanto el color como el marcador se puede indicar en la función `plot`

## Varios Subgráficos con Subplot

```python
import numpy as np
import matplotlib.pyplot as pl

# Se crea un array de numpycon con 50 elementos, empezando por el -10 y terminando en el 10
x = np.linspace(-10, 10, 50) 
diccionario = {'x': x, 'seno': np.sin(x), 'coseno': np.cos(x)}

pl.subplot(3, 2, 1)
pl.plot(diccionario['x'], diccionario['seno'], label = "Seno")
pl.legend()

pl.subplot(3, 2, 2)
pl.plot(diccionario['x'], diccionario['coseno'])

pl.subplot(3, 2, 3)
pl.plot(diccionario['x'], diccionario['seno']/diccionario['coseno'])
pl.xlim(-5,5)

pl.subplot(3, 2, 4)
pl.plot(diccionario['x'], 1/diccionario['seno'])

pl.subplot(3, 2, 5)
pl.plot(diccionario['x'], 1/diccionario['coseno'])

pl.subplot(3, 2, 6)
pl.plot(diccionario['x'], diccionario['coseno']/diccionario['seno'])
pl.xlabel('radianes')
pl.ylabel('cotag(x)')

pl.tight_layout()
pl.show()
```

## Scatter

```python
import numpy as np
import matplotlib.pyplot as pl

altura = np.array([180, 185, 160, 165, 190, 158, 172, 182, 164, 179])
peso = np.array([80, 65, 90, 81, 71, 68, 89, 91, 81, 78])

pl.scatter(altura, peso)
pl.title('Pesos y Alturas')
pl.xlabel('Altura')
pl.ylabel('Peso')
pl.show()
```

```python
import numpy as np
import matplotlib.pyplot as pl

altura = np.array([180, 185, 160, 165, 190, 158, 172, 182, 164, 179])
peso = np.array([80, 65, 90, 81, 71, 68, 89, 91, 81, 78])
sexo = np.array([0, 1, 1, 1, 1, 0, 0, 1, 1, 1])

pl.scatter(altura, peso, c=sexo)
pl.title('Pesos y Alturas')
pl.xlabel('Altura')
pl.ylabel('Peso')
pl.show()
```

## Diagrama de Tarta

```python
import numpy as np
import matplotlib.pyplot as pl

pelo = np.array(['Rubio', 'Castaño', 'Moreno', 'Pelirojo', 'Cano'])
porcentaje = np.array([15, 40, 20, 5, 20])
pl.figure(figsize =(8, 6))
pl.pie(porcentaje, labels = pelo)
pl.show()
```

## Gráficos Tridimensionales

```python
import numpy as np
import matplotlib.pyplot as pl

altura = np.array([180, 185, 160, 165, 190, 158, 172, 182, 164, 179])
peso = np.array([80, 65, 90, 81, 71, 68, 89, 91, 81, 78])
edad = np.array([15, 22, 31, 18, 52, 23, 71, 64, 81, 19])

datos = [altura,peso,edad ]

pl.matshow(datos)
pl.colorbar()
pl.show()
```
